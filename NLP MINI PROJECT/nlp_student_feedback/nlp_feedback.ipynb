```python
import pandas as pd

# Create sample dataset
data = {
    "Student_ID": [1, 2, 3, 4, 5, 6, 7, 8],
    "Feedback": [
        "The teacher explains concepts very clearly and makes class interesting.",
        "Labs are not working properly and the systems are very slow.",
        "I love the way the course is organized and how assignments are given.",
        "The lectures are too long and sometimes boring.",
        "Faculty members are helpful but the internet connection in lab is poor.",
        "The practical sessions were interactive and enjoyable.",
        "Project submission portal often crashes, please fix it soon.",
        "Everything was well managed and I learned a lot!"
    ]
}

# Create dataframe
df = pd.DataFrame(data)

# Save to CSV
df.to_csv("student_feedback.csv", index=False)

# Display the table
df
```

##  Step 3 ‚Äì Text Preprocessing

###  Theory
Raw text data often contains noise such as punctuation, capitalization, and stopwords.  
Before performing NLP operations like sentiment analysis or keyword extraction, it is important to **clean and normalize** the text.

Text preprocessing converts unstructured feedback into a form suitable for analysis.  
This process generally includes:
- **Lowercasing:** Makes all words uniform.  
- **Removing Punctuation:** Eliminates symbols that don‚Äôt add meaning.  
- **Tokenization:** Splits text into smaller units (words/tokens).  
- **Stopword Removal:** Removes common, meaningless words (e.g., ‚Äúis‚Äù, ‚Äúthe‚Äù, ‚Äúand‚Äù).  
- **Lemmatization:** Reduces words to their base form (e.g., ‚Äústudying‚Äù ‚Üí ‚Äústudy‚Äù).  

### Objective
To clean and preprocess the student feedback text so that it can be effectively used for sentiment analysis.

### Tools Used
- **NLTK:** For stopword removal and lemmatization.  
- **Regular Expressions (re):** For punctuation and character cleaning.


```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Initialize lemmatizer and stopwords
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def clean_text(text):
    # 1. Lowercase
    text = text.lower()
    # 2. Remove punctuation/special chars
    text = re.sub(r'[^a-z\s]', '', text)
    # 3. Tokenize
    tokens = nltk.word_tokenize(text)
    # 4. Remove stopwords
    tokens = [word for word in tokens if word not in stop_words]
    # 5. Lemmatize
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    # 6. Rejoin into cleaned string
    cleaned = ' '.join(tokens)
    return cleaned

# Apply preprocessing
df["Cleaned_Feedback"] = df["Feedback"].apply(clean_text)
df.head()
```

## üß© Step 4 ‚Äì Sentiment Analysis

### üß† Theory
**Sentiment Analysis** is the process of determining whether a piece of text expresses a *positive*, *negative*, or *neutral* emotion.  
In this project, we apply two popular NLP tools for sentiment detection:

1. **TextBlob:**  
   - Based on rule-based and lexical methods.  
   - Provides two scores:  
     - **Polarity** ‚Üí value between -1 (negative) and +1 (positive)  
     - **Subjectivity** ‚Üí value between 0 (objective) and 1 (subjective)

2. **VADER (Valence Aware Dictionary and sEntiment Reasoner):**  
   - A lexicon and rule-based sentiment analysis tool tuned for social-media text.  
   - Returns four scores: *positive, negative, neutral,* and *compound*.

### üéØ Objective
To analyze each feedback comment and assign a sentiment score and label.

### üß∞ Tools Used
- **TextBlob** ‚Äì For quick polarity & subjectivity.  
- **VADER (nltk.sentiment.vader)** ‚Äì For compound sentiment classification.  




```python
pip install textblob vaderSentiment
```

```python
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Initialize VADER analyzer
vader = SentimentIntensityAnalyzer()

# --- TEXTBLOB ANALYSIS ---
df["Polarity"] = df["Cleaned_Feedback"].apply(lambda x: TextBlob(x).sentiment.polarity)
df["Subjectivity"] = df["Cleaned_Feedback"].apply(lambda x: TextBlob(x).sentiment.subjectivity)

# --- VADER ANALYSIS ---
df["Vader_Compound"] = df["Cleaned_Feedback"].apply(lambda x: vader.polarity_scores(x)["compound"])

# --- SENTIMENT LABEL based on Vader compound score ---
def get_sentiment_label(score):
    if score >= 0.05:
        return "Positive"
    elif score <= -0.05:
        return "Negative"
    else:
        return "Neutral"

df["Sentiment_Label"] = df["Vader_Compound"].apply(get_sentiment_label)

# Show results
df[["Feedback", "Cleaned_Feedback", "Polarity", "Subjectivity", "Vader_Compound", "Sentiment_Label"]]
```

## Step 5 ‚Äì Data Visualization

###  Theory
Visualization helps interpret the results of sentiment analysis more effectively.  
By plotting charts, we can easily identify which type of sentiment dominates the dataset and what words appear most frequently.

In this step, we will:
1. **Visualize sentiment distribution** using bar and pie charts.  
2. **Generate a word cloud** of the most common words used in the feedback.

###  Objective
To represent the analyzed sentiment results graphically for better understanding.

###  Tools Used
- **Matplotlib / Seaborn:** For creating bar and pie charts.  
- **WordCloud:** For generating visual word frequency maps.


```python
!pip install seaborn wordcloud matplotlib
```

```python
# --- Import Visualization Libraries ---
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

# --- Sentiment Distribution (Bar Chart) ---
plt.figure(figsize=(6,4))
sns.countplot(x="Sentiment_Label", data=df, palette="pastel")
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment Type")
plt.ylabel("Number of Feedbacks")
plt.show()

# --- Sentiment Percentage (Pie Chart) ---
plt.figure(figsize=(6,6))
df["Sentiment_Label"].value_counts().plot.pie(
    autopct="%1.1f%%",
    colors=["#8fd3f4", "#f4a261", "#90be6d"]
)
plt.title("Sentiment Percentage")
plt.ylabel("")
plt.show()

# --- Word Cloud (All Feedback Combined) ---
text_all = " ".join(df["Cleaned_Feedback"])
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(text_all)

plt.figure(figsize=(10,6))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud of Student Feedback")
plt.show()
```
